## *Required Libraries*

```python
from keras.models import load_model
from keras.preprocessing.image import img_to_array,load_img
import numpy as np
import matplotlib.pyplot as plt
```

## *Classes in Every Crop*

```python
apple_classes=['Apple Scab', 'Black Rot', 'Cedar Apple Rust', 'Healthy']
cherry_classes=['Healthy', 'Powdery Mildew']
corn_classes=['Common Rust', 'Gray Leaf Spot', 'Healthy', 'Northern Leaf Blight']
grape_classes=['Black Rot', 'Esca', 'Healthy', 'Leaf Blight']
peach_classes=['Bacterial Spot', 'Healthy']
pepper_classes=['Bacterial Spot', 'Healthy']
potato_classes=['Early Blight', 'Healthy', 'Late Blight']
strawberry_classes=['Healthy', 'Leaf Scorch']
tomato_classes=['Bacterial Spot','Early Blight','Healthy','Late Blight']
```

## *Loading Our Model*

```python
apple_model=load_model('apple_model.h5')
cherry_model=load_model('cherry_model.h5')
corn_model=load_model('corn_model.h5')
grape_model=load_model('grape_model.h5')
peach_model=load_model('peach_model.h5')
pepper_model=load_model('pepper_model.h5')
potato_model=load_model('potato_model.h5')
strawberry_model=load_model('strawberry_model.h5')
tomato_model=load_model('tomato_model.h5')
```

## *Testing Our Models*

*Apple Model Testing*

```python
image=load_img('Dataset/test/AppleScab1.JPG')
image_array=img_to_array(image)
expanded_image=np.expand_dims(image_array,axis=0)

pred=apple_model.predict(expanded_image,verbose=0)
result=apple_classes[np.argmax(pred[0])]
print(result)
plt.imshow(image)
plt.axis('off')
plt.show()
```
